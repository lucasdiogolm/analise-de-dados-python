```python
# Criando o Dataframe
df = pd.read_excel('/content/drive/MyDrive/datasets/AdventureWorks.xlsx')
```

```python
#Visualizando as 5 primeiras linhas
df.head()
```

```python
# Quantidade de linhas e colunas
df.shape
```

```python
# Veificando os tipos de dados
df.dtypes
```

```python
# Qual o valor da receita total?
df['Valor Venda'].sum()
```

```python
# Custo total
df['custo'] = df['Custo Unitário'].mul(df['Quantidade']) # Criando a coluna de custo
```

```python
df.head(1)
```

```python
# Qual o custo total?
round(df['custo'].sum(), 2) #Round = arredonda em 2 casas decimais
```

```python
#Agora que temos a receita e custo e o total, podemos achar o lucro total
#Vamos criar uma coluna de Lucro = Receita - Custo
df['Lucro'] = df['Valor Venda'] - df['custo']
```

```python
df.head(5)
```

```python
#Lucro total
round(df['Lucro'].sum(), 2)
```

```python
#Criando uma coluna com total de dias para enviar o produto
df['Tempo_Envio'] = df['Data Envio'] - df['Data Venda']
```

```python
df.head(1)
```

### **Agora queremos saber a média do tempo de envio para cada marca**

```python
#Extraindo em apenas os dias
df['Tempo_Envio'] = (df['Data Envio'] - df['Data Venda']).dt.days
```

```python
df.head(1)
```

```python
#Verificando o tipo da coluna Tempo Envio
df['Tempo_Envio'].dtype
```

```python
#Média de tempo de envio por Marca
df.groupby('Marca')['Tempo_Envio'].mean()
```

### **Missing Values**

```python
#Verificando valores faltantes
df.isnull().sum()
```

### **Saber o Lucro Por Ano e Por Marca**

```python
# Vamos agrupar por ano e por marca

df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum()
```

```python
#Resetando o Index
lucro_ano = df.groupby([df['Data Venda'].dt.year, 'Marca'])['Lucro'].sum().reset_index()
lucro_ano
```

```python
#  Qual o total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False)
```

```python
# Grafico total de produtos vendidos
df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=True).plot.barh(title = 'Total Produtos Vendidos')
plt.xlabel('Total')
plt.ylabel('Produto');
```

```python
df.groupby(df['Data Venda'].dt.year)['Lucro'].sum().plot.barh(title = 'Lucro x Ano')
plt.xlabel('Ano')
plt.ylabel('Receita');
```

```python
# Selecionando apenas as vendas de 2009
df_2009 = df[df['Data Venda'].dt.year == 2009]
df_2009.head(10)
```

```python
df_2009.groupby(df_2009['Data Venda'].dt.month)['Lucro'].sum().plot(title = ' Lucro Por Mês')
plt.xlabel('Mês')
plt.ylabel('Lucro')
```

```python
df_2009.groupby('Marca')['Lucro'].sum().plot.bar(title = 'Lucro x Marca')
plt.xlabel('Marca')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```

```python
df_2009.groupby('Classe')['Lucro'].sum().plot.bar(title = 'Lucro x Classe')
plt.xlabel('Classe')
plt.ylabel('Lucro')
plt.xticks(rotation='horizontal');
```
